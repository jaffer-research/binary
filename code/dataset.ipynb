```python
import os
import pandas as pd
import shutil
```

```python

# Path to your CSV file and the folder containing all images
csv_path = '/Users/jeff/code/dataset/dr/trainLabels_cropped.csv'
images_folder = '/Users/jeff/code/dataset/dr/resized_trained_processed'  # folder where all images are initially located

# Read the CSV file
df = pd.read_csv(csv_path)
df.head()
```

```python

# Loop through each row in the CSV
for idx, row in df.iterrows():
    filename = row['image'] + '.jpeg'  # assuming the CSV column is named 'image'
    class_label = str(row['level'])  # assuming the column with class info is 'level'
    
    # Define source file path (adjust extension if needed)
    src = os.path.join(images_folder, filename)
    
    # Create a new directory for the class if it doesn't exist
    dest_dir = os.path.join(images_folder, class_label)
    os.makedirs(dest_dir, exist_ok=True)
    
    # Define destination file path
    dest = os.path.join(dest_dir, filename)
    
    # Move the file
    shutil.move(src, dest)
```
